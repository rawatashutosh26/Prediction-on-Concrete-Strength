```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
concrete_data_columns = df.columns

X = df[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
Y = df['Strength']
```

```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
n_cols = X_norm.shape[1]
n_cols
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    
    model=Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size=0.3, random_state=42)
```

```python
model = regression_model()
```

```python
epochs = 100
model.fit(X_train, Y_train, epochs=epochs, verbose=2)
```

```python
Y_pred = model.predict(X_test)
```

```python
eval_model = model.evaluate(X_test, Y_test)
eval_model
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(Y_test, Y_pred)
mean_mse = np.mean(mse)
std_mse = np.std(mse)
print("Mean of Mean_Squared_Error: ", mean_mse)
print("Squared of Mean_Squared_Error: ", std_mse)
```

```python
total_mse = 50
epochs = 100
mse_values = []
for i in range(0, total_mse):
    X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size = 0.3, random_state=i)
    model.fit(X_train, Y_train, epochs=epochs, verbose=0)
    MSE_eval = model.evaluate(X_test, Y_test, verbose=0)
    print("MSE "+str(i+1)+': '+str(MSE_eval))
    Y_prediction = model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_prediction)
    mse_values.append(mse)
    
mean_mse = np.mean(mse)
std_mse = np.std(mse)

print("Mean of Mean Squared Error: ",str(mean_mse))
print("Standard Deviation of Mean Squared Error: ",str(std_mse))
```
