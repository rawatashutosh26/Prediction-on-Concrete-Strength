```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("concrete_data.csv")
```

```python
concrete_data_columns = df.columns

X = df[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
Y = df['Strength']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=4)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    
    model=Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
model.fit(X_train, Y_train, validation_split=0.3, epochs=50, verbose=2)
```

```python
Y_pred = model.predict(X_test)
```

```python
eval_model = model.evaluate(X_test, Y_test)
eval_model
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(Y_test, Y_pred)
mean_mse = np.mean(mse)
std_mse = np.std(mse)
print("Mean of Mean_Squared_Error: ", mean_mse)
print("Squared of Mean_Squared_Error: ", std_mse)
```

```python
total_mse = 50
epochs = 50
mse_values = []
for i in range(0, total_mse):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=i)
    model.fit(X_train, Y_train, epochs=epochs, verbose=0)
    MSE_eval = model.evaluate(X_test, Y_test, verbose=0)
    print("MSE "+str(i+1)+': '+str(MSE_eval))
    Y_prediction = model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_prediction)
    mse_values.append(mse)
    
mean_mse = np.mean(mse)
std_mse = np.std(mse)

print("Mean of Mean Squared Error: ",str(mean_mse))
print("Standard Deviation of Mean Squared Error: ",str(std_mse))
```
